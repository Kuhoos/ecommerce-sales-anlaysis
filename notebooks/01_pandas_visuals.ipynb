```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("C:/Users/KIIT0001/Downloads/superstore_sales_sample.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df[df.isnull().any(axis=1)]
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
df.head()
```

```python
df["Month"]=df["Order Date"].dt.to_period('M')
```

```python
df.head()
```

```python
print(df.duplicated())
```

```python
df["Profit Margin"]=df["Profit"]/df["Sales"]
```

```python
df.head()
```

```python
condition_df=(df["Sales"]<=0) |( df["Profit"]<0)
result_df=df.loc[condition_df]
print(result_df)
```

```python
category_sales = df.groupby("Category")["Sales"].sum()
category_profit = df.groupby("Category")["Profit"].sum()

print("Highest Sales Category:", category_sales.idxmax(), "Sales:", category_sales.max())
print("Highest Profit Category:", category_profit.idxmax(), "Profit:", category_profit.max())
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
region_profit = df.groupby("Category")["Profit"].sum()

print("Highest Sales Region:", region_sales.idxmax(), "Sales:", region_sales.max())
print("Highest Profit Region:", region_profit.idxmax(), "Sales:", region_profit.max())
```

```python
subcat_sales=df.groupby("Sub-Category")["Sales"].sum()
sorted_sales=subcat_sales.sort_values(ascending=False)
sorted_sales.head(10)
```

```python
subcat_profit=df.groupby("City")["Profit"].sum()
sorted_profit=subcat_profit.sort_values(ascending=False)
sorted_profit.head(10)
```

```python
highest_profit_segment=df.groupby("Segment")["Profit"].sum()
sorted_profit1=highest_profit_segment.sort_values(ascending=False)
sorted_profit1.head(1)
```

```python
category_sales=df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
category_sales
```

```python
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
profit_region=df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
```

```python
profit_region.plot(kind='bar')
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
monthly_sales=df.groupby("Month")["Sales"].sum().sort_values(ascending=False)
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
