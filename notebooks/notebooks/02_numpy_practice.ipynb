```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/KIIT0001/Downloads/superstore_sales_sample.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="Category")
plt.title("Count of orders by category")
plt.xlabel("Category")
plt.ylabel("Number of orders")
plt.show()
```

```python
cat_sales=df.groupby("Category")["Sales"].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=cat_sales,x="Category",y="Sales")
plt.title("Sales by category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
reg_profit=df.groupby("Region")["Profit"].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=reg_profit,x="Region",y="Profit")
plt.title("Total profit by region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
cat_profit=df.groupby("Category")["Profit"].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=cat_profit,x="Profit",y="Category")
plt.title("Total profit by category")
plt.xlabel("Profit")
plt.ylabel("Category")
plt.show()
```

```python
numeric_df=df[["Sales","Profit","Quantity","Discount"]]
corr=numeric_df.corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
