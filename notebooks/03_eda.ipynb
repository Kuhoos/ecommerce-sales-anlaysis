```python
numeric_cols=["TenureMonths","MonthlyCharges","AddOnCount","TicketsLast3M","AvgMonthlyUsageHours","DiscountRate"]
for col in numeric_cols:
    df[col]=df[col].fillna(df[col].median())
    
```

```python
cat_cols=["Gender","Region","Plan","Device","SupportLevel"]
for col in cat_cols:
    df[col]=df[col].fillna(df[col].mode(0))
```
